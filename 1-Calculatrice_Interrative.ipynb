**Introduction :**

```python
nbr1 = float(input("Entrez le premier nombre : "))
nbr2 = float(input("Entrez le deuxième nombre : "))

print(nbr1, nbr2)
```

**Opérations :**

Demandez à l'utilisateur de choisir une opération parmi les suivantes : addition, soustraction, multiplication, division.
En fonction du choix, effectuez l'opération appropriée sur les deux nombres.



```python
print("Choisissez une opération:")
print("1. Addition")
print("2. Soustraction")
print("3. Multiplication")
print("4. Division")

choix = int(input("Votre choix (1/2/3/4) : "))



#Utilisation de if pour l'opération

if choix == 1:
  result = nbr1 + nbr2
elif choix == 2:
  result = nbr1 - nbr2
elif choix == 3:
  result = nbr1 * nbr2
else:
  result = nbr1 / nbr2
```

**Affichage du résultat :**

**Boucle :**

Demandez à l'utilisateur s'il souhaite effectuer une autre opération.
Si la réponse est "oui", redemandez les nombres et l'opération, puis affichez le résultat.
Répétez ce processus tant que l'utilisateur souhaite continuer.



```python
print("Bienvenue dans la Calculatrice Interactive !")
c ="oui"
while(c=="oui") :
  nbr1 = float(input("Entrez le premier nombre : "))
  nbr2 = float(input("Entrez le deuxième nombre : "))

  print("Choisissez une opération:")
  print("1. Addition")
  print("2. Soustraction")
  print("3. Multiplication")
  print("4. Division")

  choix = int(input("Votre choix (1/2/3/4) : "))

  if choix == 1:
    result = nbr1 + nbr2
  elif choix == 2:
    result = nbr1 - nbr2
  elif choix == 3:
    result = nbr1 * nbr2
  else:
    result = nbr1 / nbr2

  print("Votre resultat: ", result)
  c = input("Voulez-vous effectuer une autre opération ? (oui/non)")

  print("Merci d'avoir utilisé la Calculatrice Interactive !")
```

**Fonctions :**

Encapsulez le code d'opération dans une fonction.
Utilisez cette fonction pour effectuer les calculs.



```python
def addition(x, y):
    return x + y

def soustraction(x, y):
    return x - y

def multiplication(x, y):
    return x * y

def division(x, y):
    if y != 0:
        return x / y
    else:
        print("Erreur : Division par zéro.")
        return None
```

```python
print("Bienvenue dans la Calculatrice Interactive !\n")
c ="oui"
while(c=="oui") :
  nbr1 = float(input("Entrez le premier nombre : "))
  nbr2 = float(input("Entrez le deuxième nombre : "))

  print("Choisissez une opération:")
  print("1. Addition")
  print("2. Soustraction")
  print("3. Multiplication")
  print("4. Division")

  choix = int(input("Votre choix (1/2/3/4) : "))

  if choix == 1:
    result = addition(nbr1, nbr2)
  elif choix == 2:
    result = soustraction(nbr1, nbr2)
  elif choix == 3:
    result = multiplication(nbr1, nbr2)
  else:
    result = division(nbr1, nbr2)

  print("Votre resultat: ", result)
  c = input("Voulez-vous effectuer une autre opération ? (oui/non)")

  print("Merci d'avoir utilisé la Calculatrice Interactive !\n")
```
